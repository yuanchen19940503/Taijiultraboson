```julia
@time ΩSBHstest = ΩgwSBH.(fres(1e-12),1e-12,"new")
```

```julia
c1,c2,c3,c4,c5,c6 = "#bb2a44" , "#d79f23" , "#3c9f6a" , "#5b0b0b" , "#e37a2b" , "#525252"
labelSize = 12;
tickSize =  14;

plot(fres(10^(-12.0))[map(x-> x>0 ,ΩSBHstest)] , ΩSBHstest[map(x-> x>0 ,ΩSBHstest)] , linestyle="-" , color=c2 , linewidth=1.2 ,  label= L"$10^{-11}$")

xlim(1.0e-9  , 1.0e-6)
# ylim(1.0e-11 , 1.0e-3)

xlim(8e-2  , 1.0e5 )
ylim(1.0e-15 , 1.0e-3)
semilogx()
semilogy()
xlabel( "f / Hz" , fontsize=18)
ylabel(L"${\Omega}_{\mathrm{GW}}(f)$", fontsize=18)
legend(fontsize=10 )
legend(title=L"\chi\in[0,1]",loc=3) 
xticks(fontsize=tickSize) 
yticks(fontsize=tickSize)
# grid(linestyle="--")
# savefig("../latex/newmodel.pdf")
show()
```
