# $\Omega_{\mathrm{PI}}$ for LSD 

```julia
function SNR_LSDPI_int0(f,β)    
#     ρ=1.0
    Tobs = 4*365*86400.0
    fref = 1.0e2
    num = (f/fref)^(2.0*β)
    den = Ωn_LSD(f)^2
    return num/den*2.0*Tobs
end

@memoize function Ωβ(β)
    fmin = 1.0e4
    fmax = 3.0e5
    
    Ωβ_tobeint(f,β) = (fmax-fmin)* SNR_LSDPI_int0(fmin + (fmax-fmin)*f ,β)
    
    function integrand(x, f)
        f[1] = Ωβ_tobeint(x[1], β)
    end
    result, err = cuhre(integrand ,  rtol=1e-3 , maxevals=Int(1e9))
    result[1]^(-0.5)
end
```

```julia
βs = range(-8.0,8.0,length=30)


ΩPI_LSD(f) = maximum(Ωβ.(βs).*(f/1.0e2).^βs)
```

```julia
PIfs = 10 .^ range(3.5,6,length=100)
loglog(PIfs , ΩPI_LSD.(PIfs) , linewidth = 2.0 , color="black")
for i in βs
    loglog(PIfs , Ωβ(i).*(PIfs./1.0e2).^i ,linewidth=1.0 , alpha=0.5,linestyle= "--")
end
xlim(5.0e3 , 1.0e6)
ylim(1.0e-4 , 1e5)
grid(linestyle="--")
xlabel("f/Hz" ,fontsize=14)
ylabel(L"\Omega_{\mathrm{PI}}(f)" ,fontsize=14)
xticks(fontsize=12) 
yticks(fontsize=12)
show()
# savefig("../latex/ΩPILSD.pdf")
backup(PIfs,ΩPI_LSD.(PIfs),"ΩPILSD.txt")
```

```julia
PIfs = 10 .^ range(log10(h_nx[1]),log10(h_nx[end]),length=1000)
function SNR_NSEMOPI_int0(f,β)    
#     ρ=1.0
    Tobs = 4*365*86400.0
    fref = 1.0e2
    num = (f/fref)^(2.0*β)
    den = (Ωn_nsemo(f))^2
    return num/den*2.0*Tobs
end

@memoize function ΩβNSEMO(β)
    fmin = h_nx[1]
    fmax = h_nx[end]
    
    Ωβ_tobeint(f,β) = (fmax-fmin)* SNR_NSEMOPI_int0(fmin + (fmax-fmin)*f ,β)
    
    function integrand(x, f)
        f[1] = Ωβ_tobeint(x[1], β)
    end
    result, err = cuhre(integrand ,  rtol=1e-3 , maxevals=Int(1e9))
    result[1]^(-0.5)
end
```

```julia
βs = range(-20.0,20.0,length=40)
ΩPI_NSEMO(f) = maximum(ΩβNSEMO.(βs).*(f/1.0e2).^βs)
```

```julia
loglog(PIfs , ΩPI_NSEMO.(PIfs) , linewidth = 2.0 , color="black")
for i in βs
    loglog(PIfs , ΩβNSEMO(i).*(PIfs./1.0e2).^i ,linewidth=1.0 , alpha=0.5,linestyle= "--")
end

xlim(24 , 1.0e4)
ylim(1.0e-11 , 1e-4)
grid(linestyle="--")
xlabel("f/Hz" ,fontsize=14)
ylabel(L"\Omega_{\mathrm{PI}}(f)" ,fontsize=14)
xticks(fontsize=12) 
yticks(fontsize=12)
show()
backup(PIfs , ΩPI_NSEMO.(PIfs) , "ΩPINSEMO.txt")
# savefig("../latex/ΩPINSEMO.pdf")
```

```julia
function SNRPBH_int0(f, ms , Mmin , Mmax ,  α , det , TObs = 4.0)
    
    if det == "LSD"
        Pn = Sn_LSD(f) / 5.0
    elseif det =="NSEMO"
        Pn = Sn_nsemo(f) / 5.0
    else
        detector0 = LIGO(name=det, TObs=TObs);
        Pn = detector0.Sn(f) / 5.0
    end
    
    Ωhs = interpolate((fres(ms),), Ωgw.(fres(ms) , ms , χmin , χmax , "new"), Gridded(Linear()))
    if 1.0*fres(ms)[1] < 1.0*f < 1.0*fres(ms)[end]
        if det == "LIGO"
            Rf = γLIGO(f)/5
        else
            Rf = 1.0/5
        end
        Ωh = Ωhs(f)
        Sh = 3.0*H0_SI^2.0 * Ωh / 2.0 /f^3.0 / π^2.0
        num = Sh^2.0 * Rf^2.0
        den = (1.0/25+Rf^2.0)*Sh^2.0+Pn^2.0 + 2.0*Sh*Pn/5.0
        Tobs = TObs*365.0*86400.0
        return num/den*Tobs
    else
        return 0
    end
end

@memoize function SNR_ms(ms  ,χmin , χmax , det , TObs = 4.0)
    if det == "LSD"
        fmin = 1.0e4
        fmax = 3.0e5
    elseif det =="NSEMO"
        fmin = h_nx[1]
        fmax = h_nx[end]
    else
        detector = LIGO(name=det, TObs=TObs);
        fmin = detector.fPlotRange[1]
        fmax = detector.fPlotRange[2]
    end
    
    SNR_ms_tobeint(f,ms  ,χmin , χmax, det , TObs) = (fmax-fmin)* SNR_int0(fmin + (fmax-fmin)*f , ms  ,χmin , χmax , det , TObs)
    
    function integrand(x, f)
        f[1] = SNR_ms_tobeint(x[1],  ms  ,χmin , χmax, det , TObs)
    end
    result, err = cuhre(integrand ,  rtol=1e-3 , maxevals=Int(1e9))
    result[1]^(0.5)
end
```
