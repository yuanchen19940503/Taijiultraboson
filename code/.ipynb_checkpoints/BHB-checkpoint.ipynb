```python
%pylab inline
import gwaxion
import time
```

## sensitivity curves for LSD

```python
fres = [1.0e4  , 2.0e4, 3.0e4  , 4.0e4  , 5.0e4,  6.0e4  , 1.0e5  , 3.0e5]
hlim = [2.9e-22, 7e-23, 4.0e-23, 2.5e-23, 2.0e-23,1.7e-23, 1.3e-23, 9e-24]
```

## Setting up parameters

```python
mass_bh = 0.1 #Msun
chi_bh  = 0.99

bh = gwaxion.BlackHole(mass_bh, chi_bh, msun=True)
distance = 1e3 * gwaxion.PC_SI
distance_kpc = distance/gwaxion.PC_SI/1.0e3
```

## The dominant $(\tilde{\ell}=2,\, \tilde{m}=2)$ mode of the strain

```python
# create an array of GW frequencies
fgws = arange(1e3, 1e6, 10)

%time h0_test= bh.h0r_fit(fgws)/distance
```

```python
# plot it
loglog(fgws, h0_test, lw=2, alpha=0.8, c='k')
plot(fres , hlim , lw=2 , c="brown" , ls ='--')

ylim(1e-30,1e-20)
xlim(1.0e4 , 3.0e5)
xlabel(r'$f_{\rm GW}$ (Hz)' ,  fontsize= 15);
ylabel(r'$h_0$' , fontsize= 15);


tick_params(axis="both", which="both", direction="in")
# legend(fontsize = 15)
title(r"GWs from $m_{\rm bh}}$=%.1f$M_\odot$, d=%.1f kpc, $\chi$=%.2f" %(mass_bh,distance_kpc,chi_bh) , fontsize=15)
xticks(fontsize=12);
yticks(fontsize=12);
show()
```

## GWs from different $m_{bh}$, fixing $\chi=0.99$ and $d=1$ kpc

```python
mbhs = np.arange(0.1 , 1.1 , 0.1)
h0_mass = ones((len(mbhs) ,len(fgws) ))


for i in range(len(mbhs)):
    bh = gwaxion.BlackHole(mbhs[i], chi_bh, msun=True)
    startTime = time.time() 
    h0_mass[i] = bh.h0r_fit(fgws)/distance
    executionTime = (time.time() - startTime)
    j = len(mbhs)-(i+1)
    etime = executionTime*j
    print('Execution time:%.4f seconds.' %executionTime  + '  %1.f point(s) left. ' % j + 'About %1.f seconds.' %etime)
```

```python
# plot it
for i in range(len(mbhs)):
    loglog(fgws, h0_mass[i], label=r'$m_{\rm bh}$=%.1f '%(mbhs[i]), lw=2, alpha=0.8)

plot(fres , hlim , lw=2 , c="brown" , ls='--')

ylim(1e-30,1e-20)
xlim(1.0e4 , 1.0e6)
xlabel(r'$f_{\rm GW}$ (Hz)' ,  fontsize= 12);
ylabel(r'$h_0$' , fontsize= 12);


tick_params(axis="both", which="both", direction="in")
legend(fontsize = 10)
title(r"GWs from $\chi$=%.2f, d=%.1f kpc BHs" %(chi_bh,distance_kpc) , fontsize=15)
xticks(fontsize=12);
yticks(fontsize=12);
# savefig("../latex/GW_mass.pdf")
show()
```

## GWs from different $\chi$, fixing $m_{\rm bh}=0.1M_\odot$ and $d=1$ kpc

```python
χs = np.arange(0.50 , 1.01 , 0.1)

χs[len(χs)-1]=0.99 # χ < 1 must be satisfied

h0_spin = ones((len(χs) ,len(fgws) ))


for i in range(len(χs)):
    bh = gwaxion.BlackHole(mass_bh, χs[i], msun=True)
    startTime = time.time() 
    h0_spin[i] = bh.h0r_fit(fgws)/distance
    executionTime = (time.time() - startTime)
    j = len(mbhs)-(i+1)
    print('Execution time:%.4f seconds.' %executionTime  + '  %1.f point(s) left. ' % j)
```

```python
# plot it
for i in range(len(χs)):
    loglog(fgws, h0_spin[i], label=r'$χ$=%.2f '%(χs[i]), lw=2, alpha=0.8)

plot(fres , hlim , lw=2 , c="brown" , ls='--')

ylim(1e-35,1e-20)
xlim(1.0e4 , 1.0e6)
xlabel(r'$f_{\rm GW}$ (Hz)' ,  fontsize= 12);
ylabel(r'$h_0$' , fontsize= 12);


tick_params(axis="both", which="both", direction="in")
legend(fontsize = 10)
title(r"GWs from $m_{\rm bh}}$=%.1f$M_\odot$, d=%.1f kpc BHs" %(mass_bh,distance_kpc) , fontsize=15)
xticks(fontsize=12);
yticks(fontsize=12);
# savefig("../latex/GW_spin.pdf")
show()
```

## GWs from different $d$, fixing $m_{\rm bh}=0.1M_\odot$ and $\chi=0.99$

```python
ds = 10 **np.arange(0 , 3.01 , 0.5) # kpc
ds_SI = 1.0e3 *ds * gwaxion.PC_SI #SI unit
h0_distance = ones((len(ds) ,len(fgws) ))

bh = gwaxion.BlackHole(mass_bh, chi_bh, msun=True)

for i in range(len(ds)):
    h0_distance[i] = bh.h0r_fit(fgws)/ds_SI[i]
```

```python
# plot it
for i in range(len(ds)):
    loglog(fgws, h0_distance[i], label=r'$d$=%.1f '%(ds[i]), lw=2, alpha=0.8)

plot(fres , hlim , lw=2 , c="brown" , ls='--')

ylim(1e-35,1e-20)
xlim(1.0e4 , 1.0e6)
xlabel(r'$f_{\rm GW}$ (Hz)' ,  fontsize= 12);
ylabel(r'$h_0$' , fontsize= 12);


tick_params(axis="both", which="both", direction="in")
legend(fontsize = 10)
title(r"GWs from $m_{\rm bh}}$=%.1f $M_\odot$, $\chi$=%.2f BHs" %(mass_bh,chi_bh) , fontsize=15)
xticks(fontsize=12);
yticks(fontsize=12);
# savefig("../latex/GW_distance.pdf")
show()
```
